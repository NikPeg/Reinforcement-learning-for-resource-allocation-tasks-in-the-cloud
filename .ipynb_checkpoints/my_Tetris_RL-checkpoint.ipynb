```python
import gym_tetris
import random
from nes_py.wrappers import JoypadSpace
import gym_tetris
from gym_tetris.actions import MOVEMENT
```

```python
class JoypadSpaceMixin(JoypadSpace):
    def step(self, action):
        n_state, reward, done, info = super().step(action)
        del info["statistics"]
        return n_state, reward, done, info
```

```python
env = gym_tetris.make('TetrisA-v0')
env = JoypadSpaceMixin(env, MOVEMENT)
states = env.observation_space.shape[0]
actions = env.action_space.n
states, actions
```

```python
episodes = 1
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        n_state, reward, done, info = env.step(env.action_space.sample())
        score+=reward
        print("info:")
        print(info)
        print()
    print('Episode:{} Score:{}'.format(episode, score))
```

```python
import numpy as np
import tensorflow
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
```

```python
def build_model(states, actions):
    model = tensorflow.keras.models.Sequential()
    model.add(Flatten(input_shape=(1, 240, 256, 3)))
    model.add(Dense(24, activation='relu'))
    model.add(Dense(24, activation='relu'))
    model.add(Dense(actions, activation='linear'))
    return model
```

```python
model = build_model(states, actions)
model.summary()
```

```python
from rl.agents import DQNAgent
from rl.policy import BoltzmannQPolicy
from rl.memory import SequentialMemory
```

```python
def build_agent(model, actions):
    policy = BoltzmannQPolicy()
    memory = SequentialMemory(limit=50000, window_length=1)
    dqn = DQNAgent(model=model, memory=memory, policy=policy, 
                  nb_actions=actions, nb_steps_warmup=10, target_model_update=1e-2)
    return dqn
```

```python
dqn = build_agent(model, actions)
dqn.compile(Adam(learning_rate=1e-3), metrics=['mae'])
dqn.fit(env, nb_steps=50000, visualize=False, verbose=1)
```

```python
scores = dqn.test(env, nb_episodes=100, visualize=False)
print(np.mean(scores.history['episode_reward']))
```
